```python
# Imports

import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("output.csv");
milis = np.zeros(43);
Mbps = np.zeros(43);
for it in range(43):
    milis[it] = int(data.iat[it,1]/1000)/1000;
    Mbps[it] = int(data.iat[it,2]/10000)/100;
data.insert(3, "Latency (ms)", milis);
data.insert(4, "Throughput (Mbps)", Mbps);
display(data);
```
