```python
# Imports

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Table

data = pd.read_csv("data.csv");
milis = np.zeros(43);
Mbps = np.zeros(43);
for it in range(43):
    milis[it] = round(data.iat[it,1]/10000)/100;
    Mbps[it] = round(data.iat[it,2]/10000)/100;
data.insert(3, "Latency (ms)", milis);
data.insert(4, "Throughput (Mbps)", Mbps);
display(data);
```

```python
# Plot

rawValues = np.transpose(data.values);

figure, axis = plt.subplots(2);
figure.suptitle("Latency (Mbps) and throughput (ms)");
axis[0].plot(rawValues[0], rawValues[3]);
axis[1].plot(rawValues[0], rawValues[4]);
```

```python
# Filter

milis = np.zeros(43);
Mbps = np.zeros(43);
milis[0] = data.iat[0,3];
Mbps[0] = data.iat[0,4];
for it in range(42):
    milis[it+1] = (7*milis[it]+data.iat[it+1,3])/8;
    Mbps[it+1] = (7*Mbps[it]+data.iat[it+1,4])/8;
data.insert(5, "Filtered latency (ms)", milis);
data.insert(6, "Filtered throughput (Mbps)", Mbps);
display(data);
```

```python
# Plot filter

rawValues = np.transpose(data.values);

figure2, axis2 = plt.subplots(2);
figure2.suptitle("Filtered Latency (Mbps) and throughput (ms)");
axis2[0].plot(rawValues[0], rawValues[5]);
axis2[1].plot(rawValues[0], rawValues[6]);
```
