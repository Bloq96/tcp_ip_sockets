```python
# Imports

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Table

data = pd.read_csv("output.csv");
micros = np.zeros(43);
Mbps = np.zeros(43);
for it in range(43):
    micros[it] = round(data.iat[it,1]/100)/10;
    Mbps[it] = round(data.iat[it,2]/10000)/100;
data.insert(3, "Latency (us)", micros);
data.insert(4, "Throughput (Mbps)", Mbps);
display(data);
```

```python
# Plot

rawValues = np.transpose(data.values);

plt.plot(rawValues[0],rawValues[3]);
plt.plot(rawValues[0],rawValues[4]);
plt.ylabel("us | Mbps");
plt.xlabel("bytes");
plt.title("Latency and Throughput");
plt.show();
```
